```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

datawith_outliers = {
    'Income': [50000, 60000, 55000, 80000, 75000, 90000, 65000, 70000, 85000, 95000,1000000]
} # Sample data with an outlier

dataoutliers = pd.DataFrame(datawith_outliers);
plt.figure(figsize=(8, 6))
sns.boxplot(x=dataoutliers['Income'])
plt.title('Box Plot of Income')
plt.xlabel('Income')
plt.show()
```
