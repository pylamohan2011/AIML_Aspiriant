```python
import pandas as pd
import numpy as np
data = {
    'age': [25, 30, np.nan, 40, 45],
    'gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'income': [50000, 60000, 55000, np.nan, 75000]
}
dataframe = pd.DataFrame(data)
print(dataframe)
```

```python
dataframe['age'].fillna(dataframe['age'].mean(), inplace=True)
```

```python
dataframe['income'].fillna(dataframe['income'].mean(), inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(y=dataframe['age'], x=dataframe['income'])
plt.title('Box Plot of Income ')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()
```

```python
import pandas as pd
import numpy as np

data_with_outliers = {
    'income': [50000, 60000, 55000, 1000000, 75000,1000000]
}
dataframe_with_outliers = pd.DataFrame(data_with_outliers)

mean_income = dataframe_with_outliers['income'].mean()
median_income = dataframe_with_outliers['income'].median()

print(f"Mean Income: {mean_income}")
print(f"Median Income: {median_income}")
dataframe_with_outliers['income_mean'] = dataframe_with_outliers['income'].fillna(dataframe_with_outliers['income'].mean())
dataframe_with_outliers['income_median'] = dataframe_with_outliers['income'].fillna(dataframe_with_outliers['income'].median())
dataframe_with_outliers
print(dataframe_with_outliers)
```
